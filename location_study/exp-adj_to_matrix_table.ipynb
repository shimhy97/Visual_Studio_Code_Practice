```python
import pandas as pd
import numpy as np

def read_graph(graph_file):
    with open(graph_file,'r') as file:
        lines = file.readlines()
        node_number = int(lines[-1].split()[0]) + 1
        graph = {}
        for i in range(node_number):
            graph.setdefault(int(i),[])
        # print(lines)
        for line in lines:
            load = line.strip("\n").split()
            # print(load)
            a,b,c = int(load[0]),int(load[1]),int(load[2])
            # print(a,b,c)

            graph[a].append([b,c])

    return graph,node_number

graph, node_number = read_graph("./graph_figure6-11.txt")

adj = [graph[i] for i in graph.keys()]
df = pd.DataFrame(np.zeros(shape=(node_number,node_number)),index=[i for i in range(node_number)],columns=[i for i in range(node_number)])
for i in range(node_number):
    for j in adj[i]:
        column = j[0]
        distance = j[1]
        # print(f"{i}번 노드와 {column}번 노드 사이의 거리는 {distance}입니다.")
        df.loc[i,column]=distance
df
```
